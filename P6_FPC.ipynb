
Ejercicios prÃ¡cticos utilizando scipy.optimize

$$ min ğ‘¥_1ğ‘¥_4(ğ‘¥_1 + ğ‘¥_2 + ğ‘¥_3)+ ğ‘¥_3 $$ 

$$ ğ‘ .ğ‘¡. ğ‘¥_1ğ‘¥_2ğ‘¥_3ğ‘¥_4 â‰¥ 25 $$

$$ ğ‘¥_1^2+ğ‘¥_2^2+ğ‘¥_3^2+ğ‘¥_4^2 = 40 $$

$$ 1 â‰¤ ğ‘¥_1, ğ‘¥_2 , ğ‘¥_3, ğ‘¥_4 â‰¤ 5 $$

$$ğ‘†ğ‘’ğ‘šğ‘–ğ‘™ğ‘™ğ‘ ğ‘–ğ‘›ğ‘–ğ‘ğ‘–ğ‘ğ‘™ ğ‘¥_0 = (1,5,5,1)$$

```python
from scipy.optimize import minimize as mini
def f(x):
    return(x[0]*x[3]*(x[0]+x[1]+x[2])+x[2])
def restric1(x): #Definicion de la primera restriccion
    return (25 - (x[0]*x[1]*x[2]*x[3]))
def restric2(x): #Definicion de la primera restriccion
    return (-40 + ((x[0]**2)+(x[1]**2)+(x[2]**2)+(x[3]**2)))

constrts = [{'type':'ineq', 'fun':restric1}, {'type':'eq', 'fun':restric2}]
sem_ini = [1,5,5,1]
x_bounds = ((1, 5),(1, 5),(1, 5),(1, 5))
solutions = mini(f, sem_ini, method='SLSQP', bounds=x_bounds,constraints=constrts)
solutions
```

2. Resuelva con Minimize
$$min f = âˆ’ (2ğ‘¥_1ğ‘¥2 + 2ğ‘¥_1 âˆ’ ğ‘¥_1^2 âˆ’ 2ğ‘¥_2^2)$$

$$ ğ‘ .ğ‘¡. ğ‘¥_1^3 âˆ’ ğ‘¥_2 = 0 $$
$$ğ‘¥_2 âˆ’ (ğ‘¥_1 âˆ’ 1)^4âˆ’ 2 â‰¥ 0 $$
$$ 0.5 â‰¤ ğ‘¥_1 â‰¤ 1.5 $$
$$ 1.5 â‰¤ ğ‘¥_2 â‰¤ 2.5 $$
$$ ğ‘†ğ‘’ğ‘šğ‘–ğ‘™ğ‘™ğ‘ ğ‘–ğ‘›ğ‘–ğ‘ğ‘–ğ‘ğ‘™ ğ‘¥_0 = (0, 2.5) $$

```python
from scipy.optimize import minimize as mini
def f(x):
    return(-((2*x[0]*x[1])+(2*x[0])-(x[0]**2)-(2*x[1]**2)))
def restric1(x): #Definicion de la primera restriccion
    return ((x[0]**3-x[1]))
def restric2(x): #Definicion de la primera restriccion
    return (x[1]-((x[0]-1)**4)-2)

constrts = [{'type':'eq', 'fun':restric1}, {'type':'ineq', 'fun':restric2}]
sem_ini = [0,2.5]
x_bounds = ((0.5, 1.5),(1.5, 2.5))
solutions = mini(f, sem_ini, method='SLSQP', bounds=x_bounds,constraints=constrts)
solutions
```

$s_0$ es la cantidad de inventario que hay del prodcuto que se estÃ¡ vendiendo, puede fijar ese valor como prefiera.

Problema de macimiaxÃ³n del retorno de venta (Ingreso, Utilidad)
Precio en funcion del tiempo y la demanda:

$$ p(x_t,t)=(A-Bx_t) (D/(D+t))$$

Demanda en funciÃ³n del precio y el tiempo:
$$x(p_t,t)=(1/B)(A-p_t((D+t)/D)$$
A,B,D son parÃ¡metros que resultan de la caracterizaciÃ³n del comportamiento de ingreso como producto del precio por la demanda a lo largo del tiempo. Puede darles el valor que considere conveniente, se recomiendan A=200, B=10, D=10. 


```python
# Ejercicio4_Demanda constante
import numpy as np
D=10
A=200
B=10
t=np.linspace(1,10,10)
x=np.linspace(20,2,10)
p=0*np.linspace(1,1,10)
s0=112

from scipy.optimize import minimize as mini
def f(x):
    return(-sum(((A-B*x)*(D/(D+t)))*x))
def restric1(x): #Definicion de la primera restriccion
    return (-sum(x)+s0)
#def restric2(x): #Definicion de la primera restriccion
#    return (all(x<=A/B) and all(x>=0))

constrts = [{'type':'ineq', 'fun':restric1}]
sem_ini = p
x_bounds = ((0, A/B),(0, A/B),(0, A/B),(0, A/B),(0, A/B),(0, A/B),(0, A/B),(0, A/B),(0, A/B),(0, A/B))

solutions = mini(f, sem_ini, method='SLSQP', bounds=x_bounds,constraints=constrts)
solutions['fun']=solutions['fun']*-1
solutions
```
